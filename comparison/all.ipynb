```python
# plot free energy as a function of simulation timefor i in range(700,857,30):   
 # import fes file into pandas dataset
data1=plumed.read_as_pandas("./S1_fes_856.dat")
data2=plumed.read_as_pandas("./S3_fes_538.dat")
data3=plumed.read_as_pandas("./S6_fes_2316.dat")
data4=plumed.read_as_pandas("./S8_fes_1266.dat")


 # plot fes
plt.plot(data1["D0"],data1["file.free"], label="S1") 
plt.plot(data2["D0"],data2["file.free"], label="S3") 
plt.plot(data3["D0"],data3["file.free"], label="S6") 
plt.plot(data4["D0"],data4["file.free"], label="S8") 

# labels
plt.xlabel("Distance (nm)")
plt.ylabel("free energy [kcal/mol]")
plt.legend(loc = 'upper left',ncol=4)
plt.legend(ncol=2)
plt.ylim([0,60])
plt.savefig('All.png',bbox="tight", dpi=500, transparent=False, facecolor='white' )
```
