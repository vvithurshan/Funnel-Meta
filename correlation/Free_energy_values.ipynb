```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
import scipy.stats
```

```python

# plt.figure()
y_test = np.array([-2.13, -2.52, -3.34, -4.06])
y_pred = np.array([-2.67, -3.54, -3.94,-4.66])
r2 = r2_score(y_test, y_pred)
plt.scatter(y_test, y_pred)
plt.xlabel('Free energy Experimental (Kcal/mol)')
plt.ylabel('Free energy Predicted (Kcal/mol)')

#plt.plot(np.unique(y_test), np.poly1d(np.polyfit(y_test, y_pred, 1))(np.unique(y_test)))
plt.title('Free enenrgy predicted vs Free energy experimental ')
plt.plot(y_test, y_test,color='r')
plt.text(-4,-2.2, 'R-squared = %0.2f' % r2)
##correlation
pearson = scipy.stats.pearsonr(y_test, y_pred)
spearman, p = scipy.stats.spearmanr(y_pred, y_test)
plt.text(-4, -2.4, 'Pearson-corr = %0.2f' % pearson[0] )
plt.text(-4, -2.6, "Spearman-corr = %0.2f" % spearman)
plt.text(-4, -2.8, 'P value = %0.2f' % p)

## Error bar
plt.errorbar(y_test[0], y_pred[0], yerr=0.11, fmt='o')
plt.errorbar(y_test[1], y_pred[1], yerr=0.2, fmt='o')
plt.errorbar(y_test[2], y_pred[2], yerr=0.2, fmt='o')
plt.errorbar(y_test[3], y_pred[3], yerr=0.00005, fmt='o')
plt.savefig('correlation.png',bbox="tight", dpi=500, transparent=False, facecolor='white')
plt.show()
```
